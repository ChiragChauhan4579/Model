```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
haar_data = cv2.CascadeClassifier('./haar-cascade-files/haarcascade_frontalcatface_extended.xml')
```

```python
capture = cv2.VideoCapture(0)
data = []
while True:
    flag,img = capture.read()
    if flag:
        faces = haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img, (x,y), (x+w,y+h), (255,0,255), 4)
            face = img[y:y+h,x:x+w,:]
            face = cv2.resize(face,(50,50))
            print(len(data))
            if len(data) < 200:
                data.append(face)
        cv2.imshow('result',img)
        if cv2.waitKey(2) == 27 or len(data) >= 200:
            break
            
capture.release()
cv2.destroyAllWindows()
```

```python
np.save('without_mask.npy',data)
```

```python
np.save('with_mask.npy',data)
```

```python
plt.imshow(data[0])
```

```python
with_mask = np.load('with_mask.npy')
without_mask = np.load('without_mask.npy')
```

```python
with_mask = with_mask.reshape(200,50*50*3)
without_mask = without_mask.reshape(200,50*50*3)
```

```python
with_mask.shape,without_mask.shape
```

```python
X = np.r_[with_mask,without_mask]
```

```python
labels = np.zeros(X.shape[0])
```

```python
labels[200:] = 1.0
```

```python
names = {0 : 'Mask',1 : 'Without Mask'}
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split as tts
```

```python
x_train,x_test,y_train,y_test = tts(X,labels,test_size=0.25)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
svm = SVC()
svm.fit(x_train,y_train)
```

```python
y_pred = svm.predict(x_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
haar_data = cv2.CascadeClassifier('./haar-cascade-files/haarcascade_frontalcatface.xml')
capture = cv2.VideoCapture(0)
data = []
font = cv2.FONT_HERSHEY_COMPLEX
while True:
    flag,img = capture.read()
    if flag:
        faces = haar_data.detectMultiScale(img)
        for x,y,w,h in faces:
            cv2.rectangle(img, (x,y), (x+w,y+h), (255,0,255), 4)
            face = img[y:y+h,x:x+w,:]
            face = cv2.resize(face,(50,50))
            face = face.reshape(1,-1)
            print(face.shape)
            pred = svm.predict(face)[0]
            n = names[int(pred)]
            cv2.putText(img,n,(x,y),font,1,(255,0,255),2)
            print(n)
        cv2.imshow('result',img)
        if cv2.waitKey(2) == 27:
            break
            
capture.release()
cv2.destroyAllWindows()
```
